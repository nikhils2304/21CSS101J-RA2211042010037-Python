```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np

np.random.seed(10)
# Creating dataset with four columns
data = {
    'Column A': np.random.normal(100, 20, 200),
    'Column B': np.random.normal(90, 15, 200),
    'Column C': np.random.normal(110, 25, 200),
    'Column D': np.random.normal(105, 18, 200)
}
df = pd.DataFrame(data)

sns.set(style="whitegrid")

plt.figure(figsize=(12, 7))
sns.boxplot(data=df)

plt.title('Box Plot of Data')
plt.xlabel('Columns')
plt.ylabel('Values')

plt.show()

```

```python
# Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Upload the CSV file
from google.colab import files
uploaded = files.upload()

# Step 3: Read the CSV file
# Replace 'supermarket_sales.csv' with the name of your uploaded file
df = pd.read_csv('supermarket_sales.csv')  # Make sure the filename matches your uploaded file

# Display the first few rows to understand the data
print(df.head())

# Step 4: Creating an area chart for the first 50 entries
plt.figure(figsize=(14, 7))  # Optional: Adjust figure size for better readability

# Slice the DataFrame to get the first 50 entries
top_50 = df.head(50)

# Plotting the line chart
plt.plot(top_50['Invoice ID'], top_50['Total'], marker='o', color='b', label='Total')  # Line plot

# Creating the area chart
plt.fill_between(top_50['Invoice ID'], top_50['Total'], color='b', alpha=0.3)  # Filled area

plt.xlabel('Invoice ID')  # Set x-axis label
plt.ylabel('Total')  # Set y-axis label
plt.title('Area Chart of First 50 Invoice IDs vs Total')  # Set plot title
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.grid(True, linestyle='--', alpha=0.7)  # Add gridlines for better readability
plt.tight_layout()  # Adjust layout to prevent overlapping elements
plt.show()
```

```python
# Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Upload the CSV file
from google.colab import files
uploaded = files.upload()

# Step 3: Read the CSV file
# Replace 'supermarket_sales.csv' with the name of your uploaded file
df = pd.read_csv('supermarket_sales.csv')  # Make sure the filename matches your uploaded file

# Display the first few rows to understand the data
print(df.head())

# Step 4: Creating a pie chart for the first 50 entries
plt.figure(figsize=(10, 10))  # Optional: Adjust figure size for better readability

# Slice the DataFrame to get the first 30 entries
top_50 = df.head(10)

# Plotting the pie chart
plt.pie(top_50['Total'], labels=top_50['Invoice ID'], autopct='%1.1f%%', startangle=140)  # Pie chart
plt.title('Pie Chart of Total Values for First 50 Invoice IDs')  # Set plot title
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

plt.tight_layout()  # Adjust layout to prevent overlapping elements
plt.show()
```
